Problem 1

```python
scores = [72 , 85 , 90 , 68 , 88]
print('scores:',scores)

avr = sum(scores) / len(scores)
print('average:',avr)

highest_score = max(scores)
lowest_score = min(scores)
print('lowest score:',lowest_score)
print('highest score:',highest_score)
```

Problem 2

```python
study_hours = [1 , 2 , 3 , 4 , 5]
exam_scores = [50 , 55 , 65 , 70 , 80]

print("Study Hours and Exam Scores:")
for i in range(len(study_hours)):
    h = study_hours[i]
    s = exam_scores[i]
    print(f"Input: {h} hours → Output: {s} marks")
```

Problem 3

```python
scores = [45 , 67 , 82 , 59]

def pass_fail(score):
  for i in score:
    if i >= 60:
      print(f"{i} → Pass")
    else:
      print(f"{i} → Fail")

pass_fail(scores)
```

Problem 4

```python
sizes = [500 , 800 , 1000 , 1200]
prices = [50 , 80 , 100 , 130]


print("House Sizes and Prices:")
for size, price in zip(sizes, prices):
    print(f"Size: {size} sq ft, Price: ${price}K")

#  the price increases as the size increases.
#  positive trend
positive_trend = True
for i in range(1, len(prices)):
    if prices[i] < prices[i-1]:
        positive_trend = False
        break

if positive_trend:
    print("Positive trend: Prices increase as size increases.")
else:
    print("No consistent positive trend observed.")
```

Problem 5

```python
study_hours = [2 , 4 , 6 , 8]
scores = [55 , 65 , 75 , 85]

avr_by_2 = (scores[1] - scores[0]) / (study_hours[1] - study_hours[0])
print(f"Average increase in score for every 2 hours of study: {int(avr_by_2*2)} marks")


prd_by_10 = scores[-1] + avr_by_2 * (10 - study_hours[-1])

print(f"Predicted exam score for 10 hours of study: {int(prd_by_10)} marks")
```

Problem 6

```python
scores = [35 , 58 , 62 , 74 , 89 , 45]

for i in range(len(scores)):
  if scores[i] < 50:
    print(f"{scores[i]} → Low")
  elif 50 <= scores[i] <= 70:
    print(f"{scores[i]} → Medium")
  else:
    print(f"{scores[i]} → High")
```

Problem 7

```python
ages = [18 , 20 , 22 , 24 , 26]
heights = [160 , 165 , 170 , 175 , 180]

for i in range(len(ages)):
  print(f"Age: {ages[i]} → Height: {heights[i]}")

avarage_age = sum(ages) / len(ages)
avarage_height = sum(heights) / len(heights)
print(f"Avarage age: {avarage_age}")
print(f"Avarage height: {avarage_height}")
```
